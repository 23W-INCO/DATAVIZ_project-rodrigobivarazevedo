```python
import pandas as pd

def calculate_summary_stats(dataframe):
    summary_stats = {}

    for column in dataframe.columns:
        if pd.api.types.is_numeric_dtype(dataframe[column]):
            median_value = dataframe[column].median()
            min_value = dataframe[column].min()
            max_value = dataframe[column].max()
            q1 = dataframe[column].quantile(0.25)
            q3 = dataframe[column].quantile(0.75)
            iqr = q3 - q1
            lower_bound = q1 - (1.5 * iqr)
            upper_bound = q3 + (1.5 * iqr)

            stats = {
                'median': median_value,
                'min_value': min_value,
                'max_value': max_value,
                'q1': q1,
                'q3': q3,
                'iqr': iqr,
                'lower_bound': lower_bound,
                'upper_bound': upper_bound
            }

            summary_stats[column] = stats

    return summary_stats



```

```python
summary_stats = calculate_summary_stats(df2)
summary_stats
```

```python
with open('summary_stats.json', 'w') as json_file:
    json.dump(summary_stats, json_file, indent=4)
```

```python
summary_df = pd.DataFrame.from_dict(summary_stats, orient='index')
summary_df.to_csv('summary_stats.csv', index_label='Column')
```
